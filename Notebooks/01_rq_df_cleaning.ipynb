## Experiment process rate count 

```python
df_web.columns.tolist()
```

```python
# temporary date_time column 

df_web_2 = df_web.copy()

df_web_2["date_time"] = pd.to_datetime(
    df_web["date"].astype(str) + " " + df_web["time"].astype(str)
)
```

```python
# Adding new cols
group_cols = ["client_id", "visitor_id", "visit_id"]


df_summary = (
    df_web_2
    .groupby(group_cols)
    .agg(
        steps_count=("process_step", "count"),
        unique_steps=("process_step", "nunique"),
        finish_time=("date_time", lambda x: x[df.loc[x.index, "process_step"] == "confirm"].max())
    )
    .reset_index()
)
```

```python
# Removing nulls from time col and adding completion status

df_summary["completion_status"] = df_summary["finish_time"].notna()
df_summary["completion_status"] = df_summary["completion_status"].map(
    {True: "completed", False: "not_completed"}
)
```

```python
df_completed = df_summary[df_summary["finish_time"].notna()].copy()
```

```python
# Adding client last step column

last_steps = (
    df
    .sort_values("date_time")
    .groupby(["client_id", "visitor_id", "visit_id"])
    .last()
    .reset_index()[["client_id", "visitor_id", "visit_id", "process_step"]]
    .rename(columns={"process_step": "drop_off_step"})
)

df_summary = df_summary.merge(
    last_steps,
    on=["client_id", "visitor_id", "visit_id"],
    how="left"
)

df_summary.loc[
    df_summary["completion_status"] == "completed",
    "drop_off_step"
] = None
```

```python
# Adding average time per step each client took in the experiment

df_web_2 = df_web_2.sort_values(
    ["client_id", "visitor_id", "visit_id", "date_time"]
)

df_web_2["step_duration_sec"] = (
    df_web_2
    .groupby(["client_id", "visitor_id", "visit_id"])["date_time"]
    .diff()
    .dt.total_seconds()
)

avg_step_time = (
    df_web_2
    .groupby(["client_id", "visitor_id", "visit_id"])["step_duration_sec"]
    .mean()
    .reset_index()
    .rename(columns={"step_duration_sec": "avg_time_per_step"})
)
```

```python
df_summary = df_summary.merge(
    avg_step_time,
    on=["client_id", "visitor_id", "visit_id"],
    how="left"
)
```

```python
df_summary = df_summary.drop(
    columns=["drop_off_step_x", "drop_off_step_y"]
)
```

```python

df_web_2["date_time"] = pd.to_datetime(df_web_2["date_time"])
df_web_2 = df_web_2.sort_values(
    ["client_id", "visitor_id", "visit_id", "date_time"]
)

group_cols = ["client_id", "visitor_id", "visit_id"]


df_summary = (
    df_web_2
    .groupby(group_cols)
    .agg(
        steps_count=("process_step", "count"),
        finish_time=("date_time", lambda x: x[df_web_2.loc[x.index, "process_step"] == "confirm"].max())
    )
    .reset_index()
)

# Completion status
df_summary["completion_status"] = np.where(
    df_summary["finish_time"].notna(),
    "completed",
    "uncompleted"
)

# Client drop off step 
last_step = (
    df_web_2
    .groupby(group_cols)["process_step"]
    .last()
    .reset_index(name="last_step")
)

df_summary = df_summary.merge(
    last_step,
    on=group_cols,
    how="left",
    validate="one_to_one"
)

df_summary["drop_off_step"] = np.where(
    df_summary["completion_status"] == "completed",
    None,
    df_summary["last_step"]
)

df_summary.drop(columns="last_step", inplace=True)

# Average time per step (minutes only)

time_per_journey = (
    df_web_2
    .groupby(group_cols)["date_time"]
    .agg(lambda x: (x.max() - x.min()).total_seconds() / 60)
    .reset_index(name="total_time_minutes")
)

df_summary = df_summary.merge(
    time_per_journey,
    on=group_cols,
    how="left"
)

df_summary["avg_time_per_step"] = (
    df_summary["total_time_minutes"] / df_summary["steps_count"]
).round(2)

df_summary.drop(columns="total_time_minutes", inplace=True)

# Average time per step (minutes:seconds)
def minutes_to_mmss(x):
    if pd.isna(x):
        return None
    seconds = int(round(x * 60))
    return f"{seconds // 60}:{seconds % 60:02d}"

df_summary["avg_time_per_step_mmss"] = df_summary["avg_time_per_step"].apply(minutes_to_mmss)


df_summary["finished"] = df_summary["completion_status"] == "completed"
```
